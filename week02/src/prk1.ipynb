# Langkah 1 - Load Data

```python
import pandas as pd

data = '\semester 5\pembelajaran mesin\week02\docs\Titanic-Dataset.csv' # path dataset
df = pd.read_csv(data) # load dataset
```

# Langkah 3 - Imputasi

```python
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```
